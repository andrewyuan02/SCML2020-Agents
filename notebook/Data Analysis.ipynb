# Function to... 

```python
def get_score(file_location, last_step):
    data = pd.read_csv(file_location)
    df = data[[col for col in data.columns if col.startswith("score")]].reset_index()
    df = df[df['index'] == last_step]
    df = df.drop(['index'], axis=1)
    return df.T
```

```python
data = pd.concat([get_score('/Users/enriqueareyan/Documents/workspace/scml2020/stats' + str(i) + '.csv', last_step=9) for i in range(0, 3)]).reset_index()
```
